```python
def get_class_label(file_name, lookup):
    
    # Load the annotations file
    with open(annotations_file_path, 'r') as f:
        annotations = json.load(f)

    # Map category IDs to class labels
    category_id_to_label = {}
    categories = annotations['categories']
    for category in categories:
        category_id_to_label[category['id']] = category['name']

    image_id = lookup[file_name]
    # Retrieve the main class label for the image
    annotations = annotations['annotations']
    for annotation in annotations:
        if annotation['image_id'] == image_id:
            category_id = annotation['category_id']
            main_class_label = category_id_to_label.get(category_id, 'Unknown')
            break
    else:
        main_class_label = 'Unknown'
        
    print(category_id,':',main_class_label)
    return category_id, main_class_label
        
```
