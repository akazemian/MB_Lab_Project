```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
import matplotlib.colors as mcolors

from model_features.models.utils import load_iden
from model_evaluation.eigen_analysis.tools import *
from model_evaluation.results.predicting_brain_data.tools import write_powers
```

```python
iden = load_iden(model_name='expansion', 
                 features=3000, 
                 layers=5, 
                 dataset='naturalscenes')


file_path = f'/data/atlas/.cache/pca/{iden}_components=None'
with open(file_path, 'rb') as f:
    pca = pickle.load(f)
explained_variance = pca.explained_variance_ratio_
```

```python
rcParams['figure.figsize'] = (9,9)  


i = 0
label = [f'Features = {write_powers(2)}',f'Features = {write_powers(3)}',f'Features = {write_powers(4)}',f'Features = {write_powers(5)}']
limit_list = [100,1000,10000,100000]
color_list = []



# Sample 5 colors from the 'Blues' colormap
colors = plt.cm.Blues([0.2, 0.4, 0.6, 0.8])
# Convert RGBA colors to hexadecimal format using matplotlib.colors
hex_colors = [mcolors.rgb2hex(color) for color in colors]


for c, hex_color in enumerate(hex_colors):
    color_list.append(hex_color)


for feature in [3,30,300,3000]:
    iden = load_iden(model_name='expansion', 
                 features=feature, 
                 layers=5, 
                 dataset='naturalscenes')

    file_path = f'/data/atlas/.cache/pca/{iden}_components=None'
    with open(file_path, 'rb') as f:
        pca = pickle.load(f)
    plot_eigspec(pca.explained_variance_, label[i], color_list[i])
    i += 1

plt.ylim(0.0000001,0)
plt.xticks(size=15)
plt.yticks(size=15)
plt.legend(frameon=False,fontsize=12)
plt.savefig(f'eig_spec_comparison.png', bbox_inches='tight', dpi=100) 
```
